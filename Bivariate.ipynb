```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    #calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```

# Reject Null hypothesis p<0.05
#accept Null Hypothesis and Reject Alternate hypothesis
ANAVO:Analysis of Variance

```python
import scipy.stats as stats
stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
```

```python
#H0-there is no significant p<0.05
#H1-there is  significant difference these column
```

```python
#TWO WAY ANAVO -the means of the first factor are equal,the means of the second factor are equal and there is no interaction between the two factor
#H0 population means of first factor (factorA) are equal.population means of second factor (factorB)are equal . -Null
#H1 Atleast one of the population means of the first factor (factorA) is different.and atleast one of the population means of the
#second factor (factorB)is different

```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#perform two way ANOVA 
model = ols('param ~ ssc_p + hsc_p + degree_p',data=dataset).fit()
anova_table = sm.stats.anova_1m(model,type=2)
print(anova_table)
```
